```python
"""
to create a dataframe from a dictionary and display it
"""
import pandas as pd

arr=pd.DataFrame({'Name':['Malay','Deva','Hitesh','Tanvi'],'Semester':[5,5,4,6]})
print(arr)
```

```python
"""
to sort the DataFrame first by 'name' in ascending order
"""

import pandas as pd

arr=pd.DataFrame({'Name':['Malay','Deva','Hitesh','Tanvi','Daksh'],'Semester':[5,5,4,6,5]})
print("Before sorting")
print("\n",arr)
sort_arr=arr.sort_values(by=['Name'],ascending=True)
print("\n After sorting")
print(sort_arr)
```

```python
"""
to delete the one specific column from the DataFrame

"""

import pandas as pd

arr=pd.DataFrame({'Name':['Malay','Deva','Hitesh','Tanvi','Daksh'],'Semester':[5,5,4,6,5],'Enrollment':[85,63,83,100,76]})
print(arr)
del arr['Semester']

print("\nAfter deleting semester")
print("\n",arr)


```

```python
"""
 to write a DataFrame to CSV file using tab separator
"""
import pandas as pd

arr=pd.DataFrame({'Name':['Malay','Deva','Hitesh','Tanvi','Daksh'],'Semester':[5,5,4,6,5],'Enrollment':[85,63,83,100,76]})
print(arr)

arr.to_csv("/kaggle/working/semester.csv", sep=",",index=False)
```
