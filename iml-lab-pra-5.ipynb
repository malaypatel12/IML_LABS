```python
import numpy as np

""""
Name:- bavarva malay
Enrollment:-216270307085
to compute the mean, standard deviation, and variance of a given array along the 
second axis
"""

arr=np.array([[25,50,75,100],[125,150,175,200]])

print(arr)

print(np.mean(arr))
print("\nMean along with axis=0 ",np.mean(arr,axis=0))
print("Mean along with axis=1 ",np.mean(arr,axis=1))

print(np.std(arr))
print("\nstandard deviation along with axis 1",np.std(arr,axis=1))
print("standard deviation along with axis 0",np.std(arr,axis=0))

print("\nvariance along with axis=0",np.var(arr,axis=0))
print("variance along with axis=1",np.var(arr,axis=1))

print("\n",np.var(arr))
```
