```python
"""
Name:- Bavarva Malay
Enrollment :- 216270307085

"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../input/nifty50-stock-market-data/ADANIPORTS.csv')

data['Date'] = pd.to_datetime(data['Date'])

start_date = '2018-01-01'
end_date = '2018-01-31'

filtered_data = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]

plt.figure(figsize=(10, 5))
plt.plot(filtered_data['Date'], filtered_data['Open'], label='Opening Price', color='red', marker='o')
plt.plot(filtered_data['Date'], filtered_data['Close'], label='Closing Price', color='green', marker='x')

plt.title('Opening and Closing Stock Prices of AdaniPorts Bank')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
